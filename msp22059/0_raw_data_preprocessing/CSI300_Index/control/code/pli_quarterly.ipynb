```python
import pandas as pd
import numpy as np
```

```python
df_index= pd.read_excel(r"data/CSI300/CSI300_Index_Level.xlsx")
df_index.head()
```

```python
df_2008 = pd.read_excel(r"data/CSI300/CSI300_Index_Level(2008-3-31).xlsx")
df_2008["指标名称"][3094]
index0 = df_2008["沪深300指数"][3094]
```

```python
diff = np.array(df_index["index_level"])
diff = np.diff(diff)
diff0 = df_index["index_level"][0] - index0
diff = np.insert(diff, 0, diff0)
```

```python
is_rising = []
df_index['diff'] = diff.tolist()
rising_value = -1
for x in df_index['diff']:
    if x >0:
        rising_value = 1
    else:
        rising_value = 0
    
    is_rising.append(rising_value)


df_index['is_rising'] = is_rising
```

```python
df_index['date']=pd.to_datetime(df_index['date'])
df_index.set_index("date", inplace=True)
```

```python
last = df_index.loc[df_index.groupby(df_index.index.to_period('Q')).apply(lambda x: x.index.max())]
first = df_index.loc[df_index.groupby(df_index.index.to_period('Q')).apply(lambda x: x.index.min())]
```

```python
count = -1
pli = -1
plis = []
for i in range(0, len(last)):
    count = 0
    temp = df_index[first.index[i]:last.index[i]]
    for x in temp["is_rising"]:
        if x == 1:
            count +=1

    pli = count/len(temp)
    plis.append(pli)
```

```python
last["pli"] = plis
```

```python
last['quarter'] = range(1, len(last) + 1)
last.set_index("quarter", inplace=True)
```
