```python
import pandas as pd
df_camount = pd.read_excel(r"D:\0Laura\0-HKU\1-SEM2\project\data\CSI300_Index & Future\Control Variables\CSI300_Index_TradingVolume.xlsx", engine='openpyxl')
df_camount.head()
```

```python
df_camount['date']=pd.to_datetime(df_camount['date'])
df_camount.set_index("date", inplace=True)
Monthly_groups=df_camount.groupby(pd.Grouper(freq='M'))
Monthly_camount=Monthly_groups.sum()
Monthly_camount
```

```python
#drop is_zero row
Monthly_camount = Monthly_camount[Monthly_camount['camount'] != 0]
Monthly_camount
```

```python
# make close column value natural log of close
import numpy as np
Monthly_camount['camount'] = np.log(Monthly_camount['camount'])
Monthly_camount
```

```python
#add index week1 week2 week3...
Monthly_camount['month'] = range(1, len(Monthly_camount) + 1)
Monthly_camount
```

```python
Monthly_camount.set_index("month", inplace=True)
Monthly_camount
```

```python
#to csv
Monthly_camount.to_csv('camount_monthlyly.csv')
```
