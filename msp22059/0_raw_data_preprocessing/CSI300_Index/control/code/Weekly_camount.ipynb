```python
df_camount['date']=pd.to_datetime(df_camount['date'])
df_camount.set_index("date", inplace=True)
weekly_groups=df_camount.groupby(pd.Grouper(freq='W'))
weekly_camount=weekly_groups.sum()
weekly_camount
```

```python
#drop is_zero row
weekly_camount = weekly_camount[weekly_camount['camount'] != 0]
weekly_camount
```
