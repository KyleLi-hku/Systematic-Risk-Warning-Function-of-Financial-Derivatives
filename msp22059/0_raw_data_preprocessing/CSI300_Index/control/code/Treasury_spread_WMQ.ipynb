```python
#read 6m and 10y
import pandas as pd
sixMonth = pd.read_excel(r"datasets\Control Variables\YTM_TBond_6m.xlsx", engine='openpyxl')
tenYear = pd.read_excel(r"datasets\Control Variables\YTM_TBond_10y.xlsx", engine='openpyxl')
```

```python
sixMonth
```

```python
tenYear
```

```python
#use ten year minus six month to get the spread
spread = tenYear['YTM_10y'] - sixMonth['YTM_6m']
#convert spread to dataframe
spread = pd.DataFrame(spread)
spread.reset_index()
spread
```

```python
spread['date']= sixMonth['date']
spread
```

```python
spread['date'] = pd.to_datetime(spread['date'])
spread.set_index('date', inplace=True)
spread.rename(columns={0:'spread'}, inplace=True)

spread
```

```python
#calculate weekly avg spread
spread_weekly = spread.resample('W').mean()
spread_weekly
```

```python
#print out the row of null value 
spread_weekly[spread_weekly.isnull().any(axis=1)]

```

```python
#drop dates that are 2011-10-09, 2012-01-29, 2013-02-17, 2016-02-14, 2016-10-09, 2022-10-09 6 more rows
spread_weekly = spread_weekly.drop('2011-10-09')
spread_weekly = spread_weekly.drop('2012-01-29')
spread_weekly = spread_weekly.drop('2013-02-17')
spread_weekly = spread_weekly.drop('2016-02-14')
spread_weekly = spread_weekly.drop('2016-10-09')
spread_weekly = spread_weekly.drop('2022-10-09')
```

```python
spread_weekly.dropna(inplace=True)
spread_weekly
```

```python
#add week as column from 1 to 652
spread_weekly['week'] = range(1, len(spread_weekly) + 1)
#set week as index
spread_weekly.set_index('week', inplace=True)
spread_weekly
```

```python
#to csv
spread_weekly.to_csv(r'datasets\Control Variables\spread_weekly.csv')
```
