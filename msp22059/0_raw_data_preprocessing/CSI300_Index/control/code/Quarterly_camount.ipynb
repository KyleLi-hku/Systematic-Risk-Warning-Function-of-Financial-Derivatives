```python
import pandas as pd
df_camount = pd.read_excel(r"D:\0Laura\0-HKU\1-SEM2\project\data\CSI300_Index & Future\Control Variables\CSI300_Index_TradingVolume.xlsx", engine='openpyxl')
df_camount.head()
```

```python
df_camount['date']=pd.to_datetime(df_camount['date'])
df_camount.set_index("date", inplace=True)
quarterly_groups=df_camount.groupby(pd.Grouper(freq='Q'))
quarterly_camount=quarterly_groups.sum()
quarterly_camount
```

```python
#drop is_zero row
quarterly_camount = quarterly_camount[quarterly_camount['camount'] != 0]
quarterly_camount
```

```python
# make close column value natural log of close
import numpy as np
quarterly_camount['camount'] = np.log(quarterly_camount['camount'])
quarterly_camount
```

```python
#add index week1 week2 week3...
quarterly_camount['quarter'] = range(1, len(quarterly_camount) + 1)
quarterly_camount
```

```python
quarterly_camount.set_index("quarter", inplace=True)
quarterly_camount
```

```python
#to csv
quarterly_camount.to_csv('camount_quarterly.csv')
```
