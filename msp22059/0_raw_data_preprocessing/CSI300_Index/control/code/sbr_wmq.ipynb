```python
import pandas as pd
df_sbr = pd.read_excel(r"D:\0Laura\0-HKU\1-SEM2\project\data\CSI300_Index & Future\Control Variables\short_balance_ratio.xlsx", engine='openpyxl')
df_sbr.head()
```

```python
df_sbr['date']=pd.to_datetime(df_sbr['date'])
df_sbr.set_index("date", inplace=True)
weekly_groups=df_sbr.groupby(pd.Grouper(freq='W'))
monthly_groups=df_sbr.groupby(pd.Grouper(freq='M'))
quarterly_groups=df_sbr.groupby(pd.Grouper(freq='Q'))

weekly_sbr=weekly_groups.mean()
monthly_sbr=monthly_groups.mean()
quarterly_sbr=quarterly_groups.mean()
```

```python
weekly_sbr
```

```python
# check if there's missing value
wmissing_values = weekly_sbr.isnull()
print(wmissing_values.sum())
weekly_sbr = weekly_sbr.dropna()
weekly_sbr
```

```python
mmissing_values = monthly_sbr.isnull()
print(mmissing_values.sum())
monthly_sbr = monthly_sbr.dropna()
monthly_sbr
```

```python
qmissing_values = quarterly_sbr.isnull()
print(qmissing_values.sum())
quarterly_sbr = quarterly_sbr.dropna()
quarterly_sbr
```

```python
#add index
weekly_sbr['week'] = range(1, len(weekly_sbr) + 1)
monthly_sbr['month'] = range(1, len(monthly_sbr) + 1)
quarterly_sbr['quarter'] = range(1, len(quarterly_sbr) + 1)

weekly_sbr.head()
```
