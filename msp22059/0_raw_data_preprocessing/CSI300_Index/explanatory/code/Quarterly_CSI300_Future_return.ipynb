```python
import pandas as pd
df_closingPrice = pd.read_excel('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Future_ClosingPrice.xlsx')
```

```python
df_closingPrice.head()
```

```python
df_closingPrice['date']=pd.to_datetime(df_closingPrice['date'])
df_closingPrice.set_index("date", inplace=True)
Quarterly_groups=df_closingPrice.groupby(pd.Grouper(freq='Q'))
Quarterly_closingPrice=Quarterly_groups.mean()
Quarterly_closingPrice
```

```python
#add index week1 week2 week3...
Quarterly_closingPrice['Quarter'] = range(1, len(Quarterly_closingPrice) + 1)
Quarterly_closingPrice
```

```python
# make close column value natural log of close
import numpy as np
Quarterly_closingPrice['closing_price'] = np.log(Quarterly_closingPrice['closing_price'])
Quarterly_closingPrice
```

```python
#check missing value
Quarterly_closingPrice.isnull().sum()
#find out missing value
Quarterly_closingPrice[Quarterly_closingPrice.isnull().values==True]
```

```python
Quarterly_closingPrice.set_index("Quarter", inplace=True)
Quarterly_closingPrice
```

```python
#diff function
Quarterly_closingPrice['closingPrice'] = Quarterly_closingPrice['closing_price'].diff(periods=1)
Quarterly_closingPrice
```

```python
#delete closing_price column
del Quarterly_closingPrice['closing_price']
Quarterly_closingPrice
```

```python
#to csv
Quarterly_closingPrice.to_csv('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/Quarterly_CSI300_Future_Return.csv', index = False, header=True)
```
