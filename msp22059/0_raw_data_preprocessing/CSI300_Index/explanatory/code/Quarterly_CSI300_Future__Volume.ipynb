```python
import pandas as pd
volume = pd.read_excel(r"D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Volume.xlsx", engine='openpyxl')
volume.head()
```

```python
volume['date']=pd.to_datetime(volume['date'])
volume.set_index("date", inplace=True)
Quarterly_groups=volume.groupby(pd.Grouper(freq='Q'))
Q_volume=Quarterly_groups.sum()
Q_volume
```

```python
#add index M1 M2 M3...
Q_volume['quarter'] = range(1, len(Q_volume) + 1)
Q_volume
```

```python
# make volume column value natural log of volume
import numpy as np
Q_volume['volume'] = np.log(Q_volume['volume'])
Q_volume
```

```python
#find out missing value
is_zero = Q_volume['volume'] == 0
is_zero.sum()
#print out the missing value
Q_volume[is_zero]
```

```python
#drop is_zero row
Q_volume.drop(Q_volume[is_zero].index, inplace=True)
Q_volume
```

```python
Q_volume.set_index("quarter", inplace=True)
Q_volume
```

```python
#to csv
Q_volume.to_csv('Quarterly_CSI300_Future_Volume.csv')
```
