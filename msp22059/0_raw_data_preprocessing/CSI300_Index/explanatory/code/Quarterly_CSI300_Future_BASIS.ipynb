```python
import pandas as pd
df_closing= pd.read_excel('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Future_ClosingPrice.xlsx')
df_indexLevel= pd.read_excel('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Index_Level.xlsx')
```

```python
#combind two dataframes
df_closing['index_level'] = df_indexLevel['index_level']
```

```python
df_closing['BASIS']=df_closing['closing_price']-df_closing['index_level']
df_closing
```

```python
df_closing['date']=pd.to_datetime(df_closing['date'])
df_closing.set_index("date", inplace=True)
Quarterly_groups=df_closing.groupby(pd.Grouper(freq='Q'))
Quarterly_closing=Quarterly_groups.mean()
Quarterly_closing
```

```python
#add index week1 week2 week3...
Quarterly_closing['Quarter'] = range(1, len(Quarterly_closing) + 1)
Quarterly_closing
```

```python
zero_rows = Quarterly_closing[Quarterly_closing['BASIS'] == 0]
zero_rows
```

```python
Quarterly_closing.set_index("Quarter", inplace=True)
Quarterly_closing
```

```python
del Quarterly_closing['closing_price']
del Quarterly_closing['index_level']
Quarterly_closing
```

```python
#to csv
Quarterly_closing.to_csv('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/Quarterly_CSI300_Future_Basis.xlsx')
```
