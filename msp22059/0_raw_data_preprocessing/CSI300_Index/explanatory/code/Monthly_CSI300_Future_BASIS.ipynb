```python
import pandas as pd
df_closing= pd.read_excel('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Future_ClosingPrice.xlsx')
df_indexLevel= pd.read_excel('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Index_Level.xlsx')
```

```python
#combind two dataframes
df_closing['index_level'] = df_indexLevel['index_level']
```

```python
df_closing['BASIS']=df_closing['closing_price']-df_closing['index_level']
df_closing
```

```python
df_closing['date']=pd.to_datetime(df_closing['date'])
df_closing.set_index("date", inplace=True)
Monthly_groups=df_closing.groupby(pd.Grouper(freq='M'))
Monthly_closing=Monthly_groups.mean()
Monthly_closing
```

```python
#add index week1 week2 week3...
Monthly_closing['Month'] = range(1, len(Monthly_closing) + 1)
Monthly_closing
```

```python
Monthly_closing.set_index("Month", inplace=True)
Monthly_closing
```

```python
del Monthly_closing['closing_price']
del Monthly_closing['index_level']
Monthly_closing
```

```python
#to csv
Monthly_closing.to_csv('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/Monthly_CSI300_Future_BASIS.csv')
```
