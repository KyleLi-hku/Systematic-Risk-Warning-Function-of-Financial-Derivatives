```python
import pandas as pd
amount = pd.read_excel(r"D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Amount.xlsx", engine='openpyxl')
amount.head()
```

```python
amount['date']=pd.to_datetime(amount['date'])
amount.set_index("date", inplace=True)
Monthly_groups=amount.groupby(pd.Grouper(freq='M'))
M_amount=Monthly_groups.sum()
M_amount
```

```python
#add index M1 M2 M3...
M_amount['month'] = range(1, len(M_amount) + 1)
M_amount
```

```python
# make amount column value natural log of amount
import numpy as np
M_amount['amount'] = np.log(M_amount['amount'])
M_amount
```

```python
#find out missing value
is_zero = M_amount['amount'] == 0
is_zero.sum()
#print out the missing value
M_amount[is_zero]
```

```python
#drop is_zero row
M_amount.drop(M_amount[is_zero].index, inplace=True)
M_amount
```

```python
M_amount.set_index("month", inplace=True)
M_amount
```

```python
#to csv
M_amount.to_csv('Monthly_CSI300_Future_Amount.csv')
```
