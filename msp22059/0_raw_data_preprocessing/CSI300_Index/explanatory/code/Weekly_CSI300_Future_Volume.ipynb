```python
import pandas as pd
volume = pd.read_excel("D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Volume.xlsx")
```

```python
volume
```

```python
volume['date']=pd.to_datetime(volume['date'])
volume.set_index("date", inplace=True)
weekly_groups=volume.groupby(pd.Grouper(freq='W'))
weekly_volume=weekly_groups.sum()
weekly_volume
```

```python
#add index week1 week2 week3...
weekly_volume['week'] = range(1, len(weekly_volume) + 1)
weekly_volume
```

```python
weekly_volume = weekly_volume[weekly_volume['volume'] != 0]
weekly_volume
```

```python
# make volume column value natural log of volume
import numpy as np
weekly_volume['volume'] = np.log(weekly_volume['volume'])
weekly_volume
```

```python
weekly_volume.set_index("week", inplace=True)
weekly_volume
```

```python
weekly_volume.to_csv("Weekly_CSI300_Future_Volume.csv")
```
