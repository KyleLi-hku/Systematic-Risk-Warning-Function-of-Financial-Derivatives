```python
import pandas as pd
df_closingPrice = pd.read_excel('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Future_ClosingPrice.xlsx')
```

```python
df_closingPrice.head()
```

```python
df_closingPrice['date']=pd.to_datetime(df_closingPrice['date'])
df_closingPrice.set_index("date", inplace=True)
Monthly_groups=df_closingPrice.groupby(pd.Grouper(freq='M'))
Monthly_closingPrice=Monthly_groups.mean()
Monthly_closingPrice
```

```python
#add index week1 week2 week3...
Monthly_closingPrice['Month'] = range(1, len(Monthly_closingPrice) + 1)
Monthly_closingPrice
```

```python
# make close column value natural log of close
import numpy as np
Monthly_closingPrice['closing_price'] = np.log(Monthly_closingPrice['closing_price'])
Monthly_closingPrice
```

```python
#check missing value
Monthly_closingPrice.isnull().sum()
#find out missing value
Monthly_closingPrice[Monthly_closingPrice.isnull().values==True]
```

```python
Monthly_closingPrice.set_index("Month", inplace=True)
Monthly_closingPrice
```

```python
#diff function
Monthly_closingPrice['closingPrice'] = Monthly_closingPrice['closing_price'].diff(periods=1)
Monthly_closingPrice
```

```python
#delete closing_price column
del Monthly_closingPrice['closing_price']
Monthly_closingPrice
```

```python
#to csv
Monthly_closingPrice.to_csv('C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/Monthly_CSI300_Future_Return.csv', index = False, header=True)
```
