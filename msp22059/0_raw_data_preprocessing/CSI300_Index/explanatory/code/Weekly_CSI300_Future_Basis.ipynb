```python
import pandas as pd
df_closing= pd.read_excel(r"C:\Users\User\Desktop\finalP\datasets\CSI300_Future_ClosingPrice.xlsx", engine='openpyxl')
df_indexLevel= pd.read_excel(r"C:\Users\User\Desktop\finalP\datasets\CSI300_index_Level.xlsx", engine='openpyxl')
```

```python
#combind two dataframes
df_closing['index_level'] = df_indexLevel['index_level']
```

```python
df_closing['BASIS']=df_closing['closing_price']-df_closing['index_level']
df_closing
```

```python
df_closing['date']=pd.to_datetime(df_closing['date'])
df_closing.set_index("date", inplace=True)
weekly_groups=df_closing.groupby(pd.Grouper(freq='W'))
weekly_closing=weekly_groups.mean()
weekly_closing
```

```python
#add index week1 week2 week3...
weekly_closing['week'] = range(1, len(weekly_closing) + 1)
weekly_closing
```

```python
weekly_closing.set_index("week", inplace=True)
weekly_closing
```

```python
del weekly_closing['closing_price']
del weekly_closing['index_level']
weekly_closing
```

```python
#to csv
weekly_closing.to_csv(r'C:\Users\User\Desktop\finalP\datasets\CSI300_BASIS_weekly.csv')
```
