```python
import pandas as pd
df_closingPrice = pd.read_excel(r"C:\Users\User\Desktop\finalP\datasets\CSI300_Future_ClosingPrice.xlsx", engine='openpyxl')
```

```python
df_closingPrice.head()
```

```python
df_closingPrice['date']=pd.to_datetime(df_closingPrice['date'])
df_closingPrice.set_index("date", inplace=True)
weekly_groups=df_closingPrice.groupby(pd.Grouper(freq='W'))
weekly_closingPrice=weekly_groups.mean()
weekly_closingPrice
```

```python
#add index week1 week2 week3...
weekly_closingPrice['week'] = range(1, len(weekly_closingPrice) + 1)
weekly_closingPrice
```

```python
# make close column value natural log of close
import numpy as np
weekly_closingPrice['closing_price'] = np.log(weekly_closingPrice['closing_price'])
weekly_closingPrice
```

```python
#check missing value
weekly_closingPrice.isnull().sum()
#find out missing value
weekly_closingPrice[weekly_closingPrice.isnull().values==True]
```

```python
#drop missing value
weekly_closingPrice.dropna(inplace=True)
```

```python
weekly_closingPrice.set_index("week", inplace=True)
weekly_closingPrice
```

```python
#diff function
weekly_closingPrice['closingPrice'] = weekly_closingPrice['closing_price'].diff(periods=1)
weekly_closingPrice
```

```python
#delete closing_price column
del weekly_closingPrice['closing_price']
weekly_closingPrice
```

```python
#to csv
weekly_closingPrice.to_csv(r'C:\Users\User\Desktop\finalP\datasets\CSI300_Future_closingPrice_weekly.csv', index = False, header=True)
```
