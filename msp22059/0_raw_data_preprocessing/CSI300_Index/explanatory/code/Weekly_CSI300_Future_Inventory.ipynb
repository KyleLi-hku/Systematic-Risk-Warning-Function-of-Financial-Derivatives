```python
import pandas as pd
inventory = pd.read_excel("C:/Users/XLL/Desktop/数据/CSI300_Index & Future_0521Final/CSI300_Index & Future/CSI300_Future_Inventory.xlsx")
```

```python
inventory
```

```python
inventory['date']=pd.to_datetime(inventory['date'])
inventory.set_index("date", inplace=True)
weekly_groups=inventory.groupby(pd.Grouper(freq='W'))
weekly_inventory=weekly_groups.sum()
weekly_inventory
```

```python
#add index week1 week2 week3...
weekly_inventory['week'] = range(1, len(weekly_inventory) + 1)
weekly_inventory
```

```python
weekly_inventory = weekly_inventory[weekly_inventory['inventory'] != 0]
weekly_inventory
```

```python
# make close column value natural log of close
import numpy as np
weekly_inventory['inventory'] = np.log(weekly_inventory['inventory'])
weekly_inventory
```

```python
weekly_inventory.set_index("week", inplace=True)
weekly_inventory
```
