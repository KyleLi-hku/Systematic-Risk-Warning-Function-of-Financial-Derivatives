```python
import pandas as pd
df = pd.read_csv('C:/Users/XLL/Desktop/CS毕设项目/300_week_1.5.csv')
df
```

```python
a = ['future_return','volume','inventory','amount','BASIS']
for i in a:
    group = df[i].agg(['mean', 'std'])
    print(group)
```

```python
from scipy.stats import ttest_ind
```

```python
for i in a:
    grouped_data = df.groupby('CCt+1')[i].agg(['mean', 'std'])
    print(grouped_data)
    mean_group_0 = grouped_data.loc[0, 'mean']
    mean_group_1 = grouped_data.loc[1, 'mean']
    std_group_0 = grouped_data.loc[0, 'std']
    std_group_1 = grouped_data.loc[1, 'std']
    t_stat, p_value = ttest_ind(df.loc[df['CCt+1'] == 0, i], df.loc[df['CCt+1'] == 1, i])
    print("Mean Difference Significance (t-test):")
    print("Mean Group 0:", mean_group_0)
    print("Mean Group 1:", mean_group_1)
    print("Standard Deviation Group 0:", std_group_0)
    print("Standard Deviation Group 1:", std_group_1)
    print("T-statistic:", t_stat)
    print("P-value:", p_value)
```
