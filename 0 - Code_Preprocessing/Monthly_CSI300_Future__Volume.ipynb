```python
import pandas as pd
volume = pd.read_excel(r"D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Volume.xlsx", engine='openpyxl')
volume.head()
```

```python
volume['date']=pd.to_datetime(volume['date'])
volume.set_index("date", inplace=True)
Monthly_groups=volume.groupby(pd.Grouper(freq='M'))
M_volume=Monthly_groups.sum()
M_volume
```

```python
#add index M1 M2 M3...
M_volume['month'] = range(1, len(M_volume) + 1)
M_volume
```

```python
# make volume column value natural log of volume
import numpy as np
M_volume['volume'] = np.log(M_volume['volume'])
M_volume
```

```python
#find out missing value
is_zero = M_volume['volume'] == 0
is_zero.sum()
#print out the missing value
M_volume[is_zero]
```

```python
#drop is_zero row
M_volume.drop(M_volume[is_zero].index, inplace=True)
M_volume
```

```python
M_volume.set_index("month", inplace=True)
M_volume
```

```python
#to csv
M_volume.to_csv('Monthly_CSI300_Future_Volume.csv')
```
