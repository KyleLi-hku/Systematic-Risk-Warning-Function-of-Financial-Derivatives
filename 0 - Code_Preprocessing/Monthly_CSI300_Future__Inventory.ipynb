```python
import pandas as pd
inventory = pd.read_excel(r"D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Inventory.xlsx", engine='openpyxl')
inventory.head()
```

```python
inventory['date']=pd.to_datetime(inventory['date'])
inventory.set_index("date", inplace=True)
Monthly_groups=inventory.groupby(pd.Grouper(freq='M'))
M_inventory=Monthly_groups.sum()
M_inventory
```

```python
#add index M1 M2 M3...
M_inventory['month'] = range(1, len(M_inventory) + 1)
M_inventory
```

```python
# make amount column value natural log of amount
import numpy as np
M_inventory['inventory'] = np.log(M_inventory['inventory'])
M_inventory
```

```python
#find out missing value
is_zero = M_inventory['inventory'] == 0
is_zero.sum()
#print out the missing value
M_inventory[is_zero]
```

```python
#drop is_zero row
M_inventory.drop(M_inventory[is_zero].index, inplace=True)
M_inventory
```

```python
M_inventory.set_index("month", inplace=True)
M_inventory
```

```python
#to csv
M_inventory.to_csv('Monthly_CSI300_Future_inventory.csv')
```
