```python
import pandas as pd
df_tradingVolume = pd.read_excel(r"C:\Users\User\Desktop\finalP\datasets\CSI300_Future_Volume.xlsx", engine='openpyxl')
df_tradingVolume.head()
```

```python
df_tradingVolume['date']=pd.to_datetime(df_tradingVolume['date'])
df_tradingVolume.set_index("date", inplace=True)
weekly_groups=df_tradingVolume.groupby(pd.Grouper(freq='W'))
weekly_tradingVolume=weekly_groups.sum()
weekly_tradingVolume
```

```python
#add index week1 week2 week3...
weekly_tradingVolume['week'] = range(1, len(weekly_tradingVolume) + 1)
weekly_tradingVolume
```

```python
# make close column value natural log of close
import numpy as np
weekly_tradingVolume['amount'] = np.log(weekly_tradingVolume['volume'])
weekly_tradingVolume
```

```python
#find out missing value
is_zero = weekly_tradingVolume['volume'] == 0
is_zero.sum()
#print out the missing value
weekly_tradingVolume[is_zero]
```

```python
#drop is_zero row
weekly_tradingVolume.drop(weekly_tradingVolume[is_zero].index, inplace=True)
weekly_tradingVolume
```

```python
weekly_tradingVolume.set_index("week", inplace=True)
weekly_tradingVolume
```

```python
del weekly_tradingVolume['volume']
#to csv
weekly_tradingVolume.to_csv(r'C:\Users\User\Desktop\finalP\datasets\CSI300_Future_TradingVolume_weekly.csv')
```
