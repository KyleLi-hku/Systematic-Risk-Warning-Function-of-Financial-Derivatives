```python
import pandas as pd
inventory = pd.read_excel(r"D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Inventory.xlsx", engine='openpyxl')
inventory.head()
```

```python
inventory['date']=pd.to_datetime(inventory['date'])
inventory.set_index("date", inplace=True)
Quarterly_groups=inventory.groupby(pd.Grouper(freq='Q'))
Q_inventory=Quarterly_groups.sum()
Q_inventory
```

```python
#add index M1 M2 M3...
Q_inventory['quarter'] = range(1, len(Q_inventory) + 1)
Q_inventory
```

```python
# make amount column value natural log of amount
import numpy as np
Q_inventory['inventory'] = np.log(Q_inventory['inventory'])
Q_inventory
```

```python
#find out missing value
is_zero = Q_inventory['inventory'] == 0
is_zero.sum()
#print out the missing value
Q_inventory[is_zero]
```

```python
#drop is_zero row
Q_inventory.drop(Q_inventory[is_zero].index, inplace=True)
Q_inventory
```

```python
Q_inventory.set_index("quarter", inplace=True)
Q_inventory
```

```python
#to csv
Q_inventory.to_csv('Quarterly_CSI300_Future_inventory.csv')
```
