```python
import pandas as pd
amount = pd.read_excel(r"D:/0Laura/0-HKU/1-SEM2/project/data/CSI300_Index & Future/CSI300_Future_Amount.xlsx", engine='openpyxl')
amount.head()
```

```python
amount['date']=pd.to_datetime(amount['date'])
amount.set_index("date", inplace=True)
Quaterly_groups=amount.groupby(pd.Grouper(freq='Q'))
Q_amount=Quaterly_groups.sum()
Q_amount
```

```python
#add index M1 M2 M3...
Q_amount['quater'] = range(1, len(Q_amount) + 1)
Q_amount
```

```python
# make amount column value natural log of amount
import numpy as np
Q_amount['amount'] = np.log(Q_amount['amount'])
Q_amount
```

```python
#find out missing value
is_zero = Q_amount['amount'] == 0
is_zero.sum()
#print out the missing value
Q_amount[is_zero]
```

```python
#drop is_zero row
Q_amount.drop(Q_amount[is_zero].index, inplace=True)
Q_amount
```

```python
Q_amount.set_index("quater", inplace=True)
Q_amount
```

```python
#to csv
Q_amount.to_csv('Quartly_CSI300_Future_Amount.csv')
```
