```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:/Users/Laura/Desktop/Quarterly_CSI300_Future_inventory.csv')
df1
```

```python
df2 = pd.read_csv("C:/Users/Laura/Desktop/CSI300_IL_quarter_ML.csv")
df2
```

```python
df2["quarter"] = df2["quarter"] - 9
df2
```

```python
import numpy as np
```

```python
merged_inventory_q = pd.merge(df1, df2, on='quarter')
merged_inventory_q["rinventory"] = np.exp(merged_inventory_q["inventory"])
merged_inventory_q
```

```python
# compute the mean and standard deviation of weekly volume
mean_inv = np.mean(merged_inventory_q["inventory"])
std_inv = np.std(merged_inventory_q["inventory"])
print("mean inventory is {} ,standard deviation is {}".format(mean_inv,std_inv))
```

```python
# compute the mean and std of volume in upward period and downward period respectively 
grouped_inventory = merged_inventory_q.groupby('CCt')['inventory'].agg(['mean', 'std'])
grouped_inventory
```

```python
# Test the significance of mean difference between normal period and falling period of stock index
import scipy.stats as stats

# divide the volume data into upward and downward period respectively
u = list(merged_inventory_q[merged_inventory_q["CCt"]==1]["rinventory"])
d = list(merged_inventory_q[merged_inventory_q["CCt"]==0]["rinventory"])

# 
t_statistic, p_value = stats.ttest_ind(u, d)

# print the result of t test
print("t statistic:", t_statistic)
print("p value:", p_value)
```
