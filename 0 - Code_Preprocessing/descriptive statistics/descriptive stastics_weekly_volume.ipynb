```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:/Users/Laura/Desktop/Weekly_CSI300_Future_Volume.csv')
df1
```

```python
df2 = pd.read_csv("C:/Users/Laura/Desktop/CSI300_IL_Weekly_ML.csv")
```

```python
df2["week"] = df2["week"] - 103
df2
```

```python
import numpy as np
merged_volume_w = pd.merge(df1, df2, on='week')
merged_volume_w["rvolume"] = np.exp(merged_volume_w["volume"])
merged_volume_w
```

```python
# compute the mean and standard deviation of weekly volume
mean_vol = np.mean(merged_volume_w["volume"])
std_vol = np.std(merged_volume_w["volume"])
print("mean volume is {} ,standard deviation is {}".format(mean_vol,std_vol))
```

```python
# compute the mean and std of volume in upward period and downward period respectively 
grouped_volume = merged_volume_w.groupby('CCt')['volume'].agg(['mean', 'std'])
grouped_volume
```

```python
# Test the significance of mean difference between normal period and falling period of stock index
import scipy.stats as stats

# divide the volume data into upward and downward period respectively
u = list(merged_volume_w[merged_volume_w["CCt"]==1]["rvolume"])
d = list(merged_volume_w[merged_volume_w["CCt"]==0]["rvolume"])

# 
t_statistic, p_value = stats.ttest_ind(u, d)

# print the result of t test
print("t statistic:", t_statistic)
print("p value:", p_value)
```
