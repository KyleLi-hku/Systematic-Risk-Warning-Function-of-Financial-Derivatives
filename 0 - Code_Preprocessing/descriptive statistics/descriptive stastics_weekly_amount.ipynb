```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:/Users/Laura/Desktop/CSI300_Future_amount_weekly.csv')
df1
```

```python
df2 = pd.read_csv("C:/Users/Laura/Desktop/CSI300_IL_Weekly_ML.csv")
```

```python
df2["week"] = df2["week"] - 103
df2
```

```python
merged_amount_w = pd.merge(df1, df2, on='week')
merged_amount_w
```

```python
import numpy as np
```

```python
# compute the mean and standard deviation of weekly volume
merged_amount_w["ramount"] = np.exp(merged_amount_w["amount"])
mean_amt = np.mean(merged_amount_w["amount"])
std_amt = np.std(merged_amount_w["amount"])
print("mean amount is {} ,standard deviation is {}".format(mean_amt,std_amt))
merged_amount_w
```

```python
# compute the mean and std of volume in upward period and downward period respectively 
grouped_amount = merged_amount_w.groupby('CCt')['amount'].agg(['mean', 'std'])
grouped_amount
```

```python
# Test the significance of mean difference between normal period and falling period of stock index
import scipy.stats as stats

# divide the volume data into upward and downward period respectively
u = list(merged_amount_w[merged_amount_w["CCt"]==1]["ramount"])
d = list(merged_amount_w[merged_amount_w["CCt"]==0]["ramount"])

# t检验
statistic, p_value = stats.ttest_ind(u, d)
# 非参检验
#statistic, p_value= stats.mannwhitneyu(u,d)

# print the result of t test
print("statistic:", statistic)
print("p value:", p_value)
```
