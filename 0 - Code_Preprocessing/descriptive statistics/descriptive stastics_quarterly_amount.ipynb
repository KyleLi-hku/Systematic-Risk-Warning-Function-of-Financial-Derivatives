```python
import pandas as pd
```

```python
df1 = pd.read_csv('C:/Users/Laura/Desktop/Quartly_CSI300_Future_Amount.csv')
df1
```

```python
df2 = pd.read_csv("C:/Users/Laura/Desktop/CSI300_IL_quarter_ML.csv")
df2
```

```python
df2["quarter"] = df2["quarter"] - 9
df2
```

```python
import numpy as np
merged_amount_q = pd.merge(df1, df2, on='quarter')
merged_amount_q["ramount"] = np.exp(merged_amount_q["amount"])
merged_amount_q
```

```python
# compute the mean and standard deviation of weekly volume
mean_amt = np.mean(merged_amount_q["amount"])
std_amt = np.std(merged_amount_q["amount"])
print("mean amount is {} ,standard deviation is {}".format(mean_amt,std_amt))
```

```python
# compute the mean and std of volume in upward period and downward period respectively 
grouped_amount = merged_amount_q.groupby('CCt')['amount'].agg(['mean', 'std'])
grouped_amount
```

```python
# Test the significance of mean difference between normal period and falling period of stock index
import scipy.stats as stats

# divide the volume data into upward and downward period respectively
u = list(merged_amount_q[merged_amount_q["CCt"]==1]["ramount"])
d = list(merged_amount_q[merged_amount_q["CCt"]==0]["ramount"])

# t检验
statistic, p_value = stats.ttest_ind(u, d)
# 非参检验
#statistic, p_value= stats.mannwhitneyu(u,d)

# print the result of t test
print("statistic:", statistic)
print("p value:", p_value)
```
